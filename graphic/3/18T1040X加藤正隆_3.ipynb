```python
#乱数作りの練習
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
md1 = np.random.rand(10) #0～1までの数字で一様乱数を10個生成する
md2 = np.random.rand(2, 5) #0～1までの数字で2次元配列の一様乱数を5個生成する
md3 = np.random.uniform(3, 5, 10) #3以上5未満で10個生成する
md4 = np.random.uniform(3, 5, (2, 10)) #2次元バージョン
md5 = np.random.randint(0, 3, 10) #0以上3未満で整数乱数を10個生成する
md6 = np.random.randint(0, 3, (2, 10)) #2次元バージョン
targets=['aa', 'bb', 'cc', 'dd', 'ee', 'ff', 'gg']
md7 = np.random.choice(targets, 7) #配列の要素の中から、ランダムに7個選択する
md8 = np.random.choice(targets, 7, replace=False) #replace=Falseで1度しか出現しなくなる、重複なし
md9 = np.random.permutation(targets) #targets配列を並び替える、targets配列は変更されない
np.random.shuffle(targets) #targets配列自体を並び替える
md10=np.random.randn(10) #標準正規乱数
print("md1=", md1)
print("md2=", md2)
print("md3=", md3)
print("md4=", md4)
print("md5=", md5)
print("md6=", md6)
print("md7=", md7)
print("md8=", md8)
print("md9=", md9)
print("targets=", targets)
print("md10=", md10)
```
