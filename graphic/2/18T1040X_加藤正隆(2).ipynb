```python
x = sales['price'] # 
X = ( x-np.mean(x) ) / np.sqrt( np.var(x, ddof=1) )
print(np.mean(X))
print(np.var(X, ddof = 1))
```

```python
y = sales['units'] # 
Y = ( y-np.mean(y) ) / np.sqrt( np.var(y, ddof=1) ) 
print(np.mean(Y))
print(np.var(Y, ddof = 1))
```
